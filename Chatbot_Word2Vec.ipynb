```python
stopword = nltk.download('stopwords')
stopwords_ind = stopwords.words('indonesian')
```

```python
lemmer = nltk.stem.WordNetLemmatizer()

def LemTokens(tokens):
   return [lemmer.lemmatize(token) for token in tokens]
   
remove_punct_dict = dict((ord(punct), None) for punct in string.punctuation)

def LemNormalize(text):
   return LemTokens(nltk.word_tokenize(text.lower().translate(remove_punct_dict)))
```

```python
nltk.download('omw-1.4')
def response(user_response):
   chatbot_response=''
   sent_token.append(user_response)

   TfidfVec = TfidfVectorizer(ngram_range=(1,1), stop_words=stopwords_ind)

   tfidf = TfidfVec.fit_transform(sent_token)
   vals = cosine_similarity(tfidf[-1], tfidf)
   idx=vals.argsort()[0][-2]
   flat = vals.flatten()
  
   flat.sort()
   req_tfidf = flat[-2]

   if(req_tfidf==0):
       chatbot_response=chatbot_response+"Maaf kami belum mengerti pertanyaan anda"
       return chatbot_response
   else:
     chatbot_response = chatbot_response+sent_token[idx]
     return chatbot_response
```

```python
user_intro = ['hai', 'halo', 'tes', 'pagi', 'siang', 'malam', 'hi']
bot_intro = ['Halo', 'Selamat Datang', 'Hai!', 'Senang berbicara dengan Anda!']

def greetings(text):
  for word in text.split():
    if word.lower() in user_intro:
      return random.choice(bot_intro)
```

```python
user_end = ['bye', 'exit', 'selesai', 'dadah', 'selamat tinggal']
bot_end = "Terima kasih telah menggunakan chatbot ini, selamat tinggal!"

def exit_bot(text):
  for word in text.split():
    if word.lower() in user_end:
      return bot_end
```

```python
flag = True

while flag:
  user_input = input()
  user_input = user_input.lower()
  if exit_bot(user_input) is not None:
    print(exit_bot(user_input))
    flag = False
  else:
    if greetings(user_input) is not None:
      print(greetings(user_input))
    else:
      print("chatbot: ",end="")
      print(response(user_input))
      sent_token.remove(user_input)
```
